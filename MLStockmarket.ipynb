```python
import yfinance as yf
```

```python
tickerName = "NovoCure Limited"

tickerSymbol = "NVCR"
```

```python
tickerName = yf.Ticker(tickerSymbol)
```

```python
tickerName = tickerName.history(period="max")
```

```python
tickerName
```

```python
tickerName.index
```

```python
tickerName.plot.line(y="Close", use_index=True)
```

```python
del tickerName["Dividends"]
del tickerName["Stock Splits"]
```

```python
"The data from tomorrow column represents the price it opens tommorrow based on the close price on the previous day."
tickerName["Tomorrow"] = tickerName["Close"].shift(-1)
```

```python
tickerName
```

```python
tickerName["Target"] = (tickerName["Tomorrow"] > tickerName["Close"]).astype(int)
tickerName
```

```python
tickerName = tickerName.loc["1990-01-01":].copy()
tickerName
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = tickerName.iloc[:-100]

test = tickerName.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]

model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
preds
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
preds
```

```python
#the float value from precision_score represents the probability percentage of the stock going up based on the prediction results. 

precision_score(test["Target"], preds) 
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
```

Back Testing starts below

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#start = 10 years of data   #step = every trading years has 250 days

#training the model backend process = take first 10 years of data to predict the 11th year of data and then take 11 years of data to predict the 12th year of data etc. 


def backtest(data, model, predictors, min_data_points=250, step=250):
    all_predictions = []
    num_data_points = data.shape[0]

    start = min_data_points  # Adjust the start parameter to the minimum required data points

    for i in range(start, num_data_points, step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i + step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
predictions = backtest(tickerName, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

Adding extra predictors to the model

```python
#we'll calculate the mean close price at the last 2 days, 5 days, 3 months, and last 4 years 
horizons = [2,5,60,250,1000]

new_predictors = []

for horizon in horizons:
    rolling_avrg = tickerName.rolling(horizon).mean()

    ratio_column = f"Close_ratio_{horizon}"
    tickerName[ratio_column] = tickerName["Close"] / rolling_avrg["Close"]

    trend_column = f"Trend_{horizon}"
    tickerName[trend_column] = tickerName.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
tickerName = tickerName.dropna(subset=tickerName.columns[tickerName.columns != "Tomorrow"])
```

```python
tickerName
```

Further Enhancing the model

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]  
    preds[preds >= .6] = 1  
    preds[preds < .6] = 0   
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(tickerName, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
def predict_next_day(data, model, predictors):
    # Get the most recent data point
    latest_data = data.iloc[-1].copy()
    
    # Create a DataFrame for the next day's data
    next_day_data = pd.DataFrame(columns=data.columns)
    next_day_data = next_day_data.append(latest_data)
    
    # Make the prediction
    prediction = predict(data, next_day_data, predictors, model)
    
    # Get the date of the next day
    next_day = latest_data.name + pd.DateOffset(days=1)
    
    # Get the opening price
    opening_price = latest_data['Open']
    
    # Determine whether the closing price will be higher or lower
    if prediction['Predictions'].values[0] == 1:
        result = "Higher than Opening price (Buy)"
    else:
        result = "Lower than Opening price (Hold)"
    
    # Output the prediction information
    print("Date:", next_day.date())
    print("Opening Price:", opening_price)
    print("Closing Price Prediction:", result)
```

```python
predict_next_day(tickerName, model, predictors)
```
